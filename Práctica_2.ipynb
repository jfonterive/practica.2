```python
import random
import math
import matplotlib.pyplot as plt
random.seed()

n=50 #número de pasos
r=1 #longitud del paso
```

Paseo unidimensional:

```python
x = 0

listax = [x]

for i in range (n):
  if random.random() < 0.5:
    r=-1*r
  x += r
  listax.append(x)

plt.plot(listax)
print("Paseo aleatorio unidimensional:", listax)
plt.show()
```

Paseo bidimensional:

```python
x=y=0

listax=[x]
listay=[y]

for i in range (n):
  angulo=random.random()*math.pi*2
  x+=r*math.cos(angulo)
  y+=r*math.sin(angulo)

  listax.append(x)
  listay.append(y)

plt.plot(listax,listay)
print("Random walk bidimensional", listax,listay)
plt.show()

```
